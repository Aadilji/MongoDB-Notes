### suppose we want only a specific column data we can put 1, ahead it , which will make 0 on all other columns and it would not be shown to us.

```python
all_Docs = collection.find({"companyName" : "aineurox"},{"Faculty":1})

for item in all_Docs:
    print(item)  
```

#### now this putting 1 ahead of a column would not affect the id col, because it is a autogenerated feild and it may have id = 1

```python
# so to deal with it we can put 0 in the id and 1 ahead of the col which we want to get.

all_Docs = collection.find({"companyName" : "aineurox"},{'_id':0,"Faculty":1}) 

for i in all_Docs:
    print(i) 
```

# Filter query

```python
# it is done with the help of '$' keyword
# '$gt' = greater than
# '$lt' = less than

a = collection.find({'GRE Score': {'$gt':'337'}})  
```

```python
for i in a :
    print(i) 
```

```python
my_query1={"GRE Score":{"$lt":"337"}}

a = collection.find(my_query1)  
```

```python
for i in a:
    print(i) 
```
